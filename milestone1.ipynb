```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()

```

```python
df.info()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.describe(include='all')
```

```python
df.to_csv("cleaned_netflix_titles.csv", index=False)
```

Milestone 1: Dataset Preparation

Steps performed:
1. Loaded dataset using pandas
2. Checked dataset structure using df.info()
3. Handled missing values using fillna()
4. Checked duplicate records
5. Converted date_added column to datetime format
6. Generated dataset summary
7. Saved cleaned dataset for further analysis


```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df['duration_number'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_type'] = df['duration'].str.replace('[0-9]', '', regex=True).str.strip()
```

```python
df['cast_count'] = df['cast'].apply(lambda x: len(str(x).split(',')))
```

```python
df['primary_country'] = df['country'].apply(lambda x: str(x).split(',')[0])
```

```python
df.to_csv("feature_engineered_netflix.csv", index=False)
```
